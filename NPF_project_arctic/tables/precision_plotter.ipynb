```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'Times New Roman'
# For changed, smote
data1 = {
    'adb': [0.7491996162197646],
    'rf_adb': [0.7309454114084204],
    'rf_lr': [0.7957771787960467],
    'rf_xgb': [0.7902348987329408],
    'gbm': [0.7963655570252793],
    'knn': [0.779420643009833],
    'lr': [0.46012706953646515],
    'mlp': [0.5739395822782868],
    'nb': [0.49687817882072327],
    'rf': [0.7608069252886818],
    'svm': [0.8140610318299272],
    'xgb': [0.7972312510647516]
}
# For unchanged, norm
data2 = {
    'adb': [0.4306437789876991],
    'rf_adb': [0.49235448580712665],
    'rf_lr': [0.3810919540229885],
    'rf_xgb': [0.4404119362243315],
    'gbm': [0.3730237154150198],
    'knn': [0.41197107081174433],
    'lr': [0.4823232323232323],
    'mlp': [0.5034570065244911],
    'nb': [0.430743644571157],
    'rf': [0.4003475616378842],
    'svm': [0.6936026936026937],
    'xgb': [0.5128161312371838]
}

df1 = pd.DataFrame.from_dict(data1, orient='index', columns=['Value']) * 100
df2 = pd.DataFrame.from_dict(data2, orient='index', columns=['Value']) * 100
# Merge the two DataFrames on common index (names)
merged_df = pd.merge(df1, df2, left_index=True, right_index=True)
merged_df['models'] = merged_df.index

# Plotting side-by-side bar plots
plt.figure()
merged_df.plot(kind='bar', width=0.9)
plt.title('Precisions of models', fontsize=40)
plt.xlabel('Models', fontsize=30)
plt.ylabel('Precision (%)', fontsize=30)
plt.xticks(rotation=45)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.ylim(0, 100)
plt.legend(['SMOTE', 'normal'], loc='upper right', fontsize=30)

fig = plt.gcf()
fig.set_size_inches(25, 15)

plt.savefig('precisions.png', dpi=300)
```
