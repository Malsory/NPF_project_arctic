```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'Times New Roman'
# For changed, smote
data1 = {
    'adb': [0.7505376344086021],
    'rf_adb': [0.7311827956989246],
    'rf_lr': [0.7935483870967742],
    'rf_xgb': [0.7913978494623656],
    'gbm': [0.7979453479034017],
    'knn': [0.7821150097465887],
    'lr': [0.4838709677419355],
    'mlp': [0.5548387096774194],
    'nb': [0.5047040125339902],
    'rf': [0.7569892473118279],
    'svm': [0.8134656012650024],
    'xgb': [0.7935483870967742]
}
# For unchanged, norm
data2 = {
    'adb': [0.43376410872904564],
    'rf_adb': [0.4863253856942496],
    'rf_lr': [0.3931276297335203],
    'rf_xgb': [0.4277699859747545],
    'gbm': [0.43838918283362727],
    'knn': [0.4223061148396719],
    'lr': [0.4827956989247311],
    'mlp': [0.46992252721565486],
    'nb': [0.4679012345679012],
    'rf': [0.41407199625993457],
    'svm': [0.40009935793068324],
    'xgb': [0.43214452681493354]
}

df1 = pd.DataFrame.from_dict(data1, orient='index', columns=['Value']) * 100
df2 = pd.DataFrame.from_dict(data2, orient='index', columns=['Value']) * 100
# Merge the two DataFrames on common index (names)
merged_df = pd.merge(df1, df2, left_index=True, right_index=True)
merged_df['models'] = merged_df.index

# Plotting side-by-side bar plots
plt.figure()
merged_df.plot(kind='bar', width=0.9)
plt.title('Recalls of models', fontsize=40)
plt.xlabel('Models', fontsize=30)
plt.ylabel('Recall (%)', fontsize=30)
plt.xticks(rotation=45)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.ylim(0, 100)
plt.legend(['SMOTE', 'normal'], loc='upper right', fontsize=30)

fig = plt.gcf()
fig.set_size_inches(25, 15)

plt.savefig('recalls.png', dpi=300)
```
