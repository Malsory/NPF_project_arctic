```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'Times New Roman'
# For changed, smote
data1 = {
    'adb': [0.8079986125563648],
    'rf_adb': [0.9081096080471731],
    'rf_lr': [0.9164203954214359],
    'rf_xgb': [0.9307665626083942],
    'gbm': [0.9165539788273852],
    'knn': [0.8342265274148136],
    'lr': [0.7225320846340617],
    'mlp': [0.7457023933402706],
    'nb': [0.7247548013256538],
    'rf': [0.9022476586888657],
    'svm': [0.9120971108988508],
    'xgb': [0.9239195282691641]
}
# For unchanged, norm
data2 = {
    'adb': [0.5782218779447962],
    'rf_adb': [0.7500887735689116],
    'rf_lr': [0.7316091788409343],
    'rf_xgb': [0.7605178209887117],
    'gbm': [0.7523063139617391],
    'knn': [0.696983404725699],
    'lr': [0.6945310340872708],
    'mlp': [0.7023900109363339],
    'nb': [0.694375624682424],
    'rf': [0.7396746492649385],
    'svm': [0.7215253932186316],
    'xgb': [0.7044217612424334]
}

df1 = pd.DataFrame.from_dict(data1, orient='index', columns=['Value']) * 100
df2 = pd.DataFrame.from_dict(data2, orient='index', columns=['Value']) * 100
# Merge the two DataFrames on common index (names)
merged_df = pd.merge(df1, df2, left_index=True, right_index=True)
merged_df['models'] = merged_df.index

# Plotting side-by-side bar plots
plt.figure()
merged_df.plot(kind='bar', width=0.9)
plt.title('Roc auc scores of models', fontsize=40)
plt.xlabel('Models', fontsize=30)
plt.ylabel('Precision (%)', fontsize=30)
plt.xticks(rotation=45)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.ylim(0, 100)
plt.legend(['SMOTE', 'normal'], loc='upper right', fontsize=30)

fig = plt.gcf()
fig.set_size_inches(25, 15)

plt.savefig('roc_auc.png', dpi=300)
```
