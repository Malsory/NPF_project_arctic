```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'Times New Roman'
# For changed, smote
data1 = {
    'adb': [0.7505376344086021],
    'rf_adb': [0.7311827956989247],
    'rf_lr': [0.7935483870967742],
    'rf_xgb': [0.7913978494623656],
    'gbm': [0.7956989247311828],
    'knn': [0.7698924731182796],
    'lr': [0.4838709677419355],
    'mlp': [0.5548387096774193],
    'nb': [0.5118279569892473],
    'rf': [0.7569892473118279],
    'svm': [0.7784946236559139],
    'xgb': [0.7935483870967742]
}
# For unchanged, norm
data2 = {
    'adb': [0.5851528384279476],
    'rf_adb': [0.6419213973799127],
    'rf_lr': [0.6637554585152838],
    'rf_xgb': [0.7030567685589519],
    'gbm': [0.6768558951965066],
    'knn': [0.6681222707423581],
    'lr': [0.74235807860262],
    'mlp': [0.6550218340611353],
    'nb': [0.7205240174672489],
    'rf': [0.6855895196506551],
    'svm': [0.6943231441048034],
    'xgb': [0.6637554585152838]
}

df1 = pd.DataFrame.from_dict(data1, orient='index', columns=['Value']) * 100
df2 = pd.DataFrame.from_dict(data2, orient='index', columns=['Value']) * 100
# Merge the two DataFrames on common index (names)
merged_df = pd.merge(df1, df2, left_index=True, right_index=True)
merged_df['models'] = merged_df.index

# Plotting side-by-side bar plots
plt.figure()
merged_df.plot(kind='bar', width=0.9)
plt.title('Accuracies of models', fontsize=40)
plt.xlabel('Models', fontsize=30)
plt.ylabel('Accuracy (%)', fontsize=30)
plt.xticks(rotation=45)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.ylim(0, 100)
plt.legend(['SMOTE', 'normal'], loc='upper right', fontsize=30)

fig = plt.gcf()
fig.set_size_inches(25, 15)

plt.savefig('accuracies.png', dpi=300)
```
