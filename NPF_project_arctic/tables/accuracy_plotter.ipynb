```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data1 = {
    'adb': 0.7024313725490197,
    'rf_adb': 0.7300392156862745,
    'rf_lr': 0.6984313725490197,
    'rf_xgb': 0.7578039215686274,
    'gbm': 0.7700392156862745,
    'knn': 0.7698039215686274,
    'lr': 0.516235294117647,
    'rf': 0.7417254901960785,
    'svm': 0.7420392156862746,
    'xgb': 0.7701176470588236
}
# Arctic
data2 = {
    'adb': 0.7303225806451613,
    'rf_adb': 0.7427956989247312,
    'rf_lr': 0.7797849462365591,
    'rf_xgb': 0.8060215053763441,
    'gbm': 0.7363440860215054,
    'knn': 0.7987096774193547,
    'lr': 0.5118279569892473,
    'rf': 0.8124731182795699,
    'svm': 0.8064516129032258,
    'xgb': 0.7823655913978494
}

# Combined
data3 = {
    'adb': 0.727300229140274,
    'rf_adb': 0.7183741424546393,
    'rf_lr': 0.7860861318214228,
    'rf_xgb': 0.8020761077997689,
    'gbm': 0.7313977556713949,
    'knn': 0.78533156113061,
    'lr': 0.5063321634026292,
    'rf': 0.7998511626619038,
    'svm': 0.818076467778447,
    'xgb': 0.7652495967546538
}

df1 = pd.DataFrame(list(data1.items()), columns=['models', 'value1'])
df2 = pd.DataFrame(list(data2.items()), columns=['models', 'value2'])
df3 = pd.DataFrame(list(data3.items()), columns=['models', 'value3'])

df1['value1'] = df1['value1'] * 100
df2['value2'] = df2['value2'] * 100
df3['value3'] = df3['value3'] * 100

# Merge the two DataFrames on common index (names)
merged_df = pd.merge(df1, df3, on='models')
merged_df = pd.merge(merged_df, df2, on='models')

colors = ['black', 'gray', 'white']
# Plotting side-by-side bar plots
plt.figure()
ax = merged_df.plot(kind='bar', x='models', width=0.9, color=colors, edgecolor='black')
plt.title('Accuracies of models', fontsize=40)
plt.xlabel('Models', fontsize=30)
plt.ylabel('Accuracy (%)', fontsize=30)
plt.xticks(rotation=45)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.ylim(0, 100)
plt.legend(['Urban', 'Combined', 'Arctic'], loc='upper right', fontsize=30)

fig = plt.gcf()
fig.set_size_inches(25, 15)

plt.savefig('accuracies.png', dpi=300)
```
