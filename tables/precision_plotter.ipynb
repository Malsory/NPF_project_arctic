```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# plt.rcParams['font.family'] = 'Times New Roman'
# Urban
data1 = {
    'adb': 0.7194813477230357,
    'rf_adb': 0.7337591925904621,
    'rf_lr': 0.7103623301172322,
    'rf_xgb': 0.7633939864744819,
    'gbm': 0.7702155054554436,
    'knn': 0.7843014661014391,
    'lr': 0.5251650069991031,
    'rf': 0.7541578823080373,
    'svm': 0.7445283139632984,
    'xgb': 0.7702790283904835
}
# Arctic
data2 = {
    'adb': 0.7289490019562008,
    'rf_adb': 0.7408179425603996,
    'rf_lr': 0.7837315387226466,
    'rf_xgb': 0.8096752425430201,
    'gbm': 0.7345562890167544,
    'knn': 0.7997157931083333,
    'lr': 0.49472213909242957,
    'rf': 0.8146191873726956,
    'svm': 0.8075436876813491,
    'xgb': 0.7807894339863746
}

# Combined
data3 = {
    'adb': 0.7273769160086732,
    'rf_adb': 0.7165735781392695,
    'rf_lr': 0.7894631311543051,
    'rf_xgb': 0.8035956755265058,
    'gbm': 0.7305381955533942,
    'knn': 0.7888812571798908,
    'lr': 0.49077901219381126,
    'rf': 0.8000908246580654,
    'svm': 0.8170164116044546,
    'xgb': 0.7648685496194394
}

df1 = pd.DataFrame(list(data1.items()), columns=['models', 'value1'])
df2 = pd.DataFrame(list(data2.items()), columns=['models', 'value2'])
df3 = pd.DataFrame(list(data3.items()), columns=['models', 'value3'])

df1['value1'] = df1['value1'] * 100
df2['value2'] = df2['value2'] * 100
df3['value3'] = df3['value3'] * 100

# Merge the two DataFrames on common index (names)
merged_df = pd.merge(df1, df3, on='models')
merged_df = pd.merge(merged_df, df2, on='models')

colors = ['black', 'gray', 'white']
# Plotting side-by-side bar plots
plt.figure()
ax = merged_df.plot(kind='bar', x='models', width=0.9, color=colors, edgecolor='black')
plt.title('Precisions of models', fontsize=40)
plt.xlabel('Models', fontsize=30)
plt.ylabel('Precision (%)', fontsize=30)
plt.xticks(rotation=45)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.ylim(0, 100)
plt.legend(['Urban', 'Combined', 'Arctic'], loc='upper right', fontsize=30)

fig = plt.gcf()
fig.set_size_inches(25, 15)

plt.savefig('precisions.png', dpi=300)
```
