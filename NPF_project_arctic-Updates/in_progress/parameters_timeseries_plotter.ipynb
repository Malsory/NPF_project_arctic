```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline
```

Day.type count

```python
# Read dataset
df = pd.read_csv(r'CC:\\Users\\Masloriy\\Desktop\\Data\\Data_arctic\\in_progress\\output_combined.csv')
# Prep the dataset
df = df.dropna()
df = df.drop(df.columns[df.columns.str.contains('Unnamed')], axis=1)
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.dayofyear
df = df.drop(df.index[-1])
df_temp = df[['day.type', 'month']].copy()
grouped_data = df.groupby(['month', 'day.type']).size().unstack(fill_value=0)
plt.rcParams['font.family'] = 'Times New Roman'
plt.figure(figsize=(10, 6))
grouped_data.plot(kind='bar', stacked=True)
plt.title('Count of day type by months')
plt.xlabel('Month')
plt.ylabel('Count')
plt.legend(title='day.type')
plt.xticks(rotation=0)
plt.savefig('CC:\\Users\\Masloriy\\Desktop\\Data\\Data_arctic\\png\\count_day.type_month.png', dpi=300)
```

NH3

```python
plt.figure(figsize=(10, 6))
plt.rcParams['font.family'] = 'Times New Roman'

# Smooth the data
x_smooth = np.linspace(grouped_data.index.min(), grouped_data.index.max(), 300)
y_smooth = make_interp_spline(grouped_data.index, grouped_data.values)(x_smooth)

plt.plot(x_smooth, y_smooth, color='tab:blue', linewidth=2)

plt.title('NH3 Concentration Over Days')
plt.xlabel('Day')
plt.ylabel('NH3 (ppb)')
plt.xticks(rotation=0)
plt.grid(True)

plt.savefig('C:\\Users\\Masloriy\\Desktop\\Data\\Data_arctic\\png\\conc_NH3_day_smooth.png', dpi=300)
plt.show()
```
